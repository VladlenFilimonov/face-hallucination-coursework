```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("../statistics/srf_train_results.csv")

# Extract data columns
epochs = df['Epoch'].astype(int)
loss_g = df['Loss_G']
psnr = df['PSNR']
ssim = df['SSIM']

# Creating a 2x3 grid of plots
fig, axs = plt.subplots(1, 3, figsize=(15, 10))

# Plotting Loss_G
axs[0].plot(epochs, loss_g, label='Loss_G', color='orange')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Loss_G')
axs[0].set_title('Loss_G over Epochs')
axs[0].legend()
axs[0].grid(True)

# Plotting PSNR
axs[1].plot(epochs, psnr, label='PSNR', color='purple')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('PSNR')
axs[1].set_title('PSNR over Epochs')
axs[1].legend()
axs[1].grid(True)

# Plotting SSIM
axs[2].plot(epochs, ssim, label='SSIM', color='brown')
axs[2].set_xlabel('Epoch')
axs[2].set_ylabel('SSIM')
axs[2].set_title('SSIM over Epochs')
axs[2].legend()
axs[2].grid(True)

plt.show()
```
