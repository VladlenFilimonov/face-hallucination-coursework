```python
import sys
sys.path.append('../')

from model import Generator
from model import Discriminator
from torchviz import make_dot
import torch
```

```python
x = torch.randn(1, 3, 32, 32)
generator = Generator()
y = generator(x)
graph = make_dot(y.mean(), params=dict(generator.named_parameters()))
graph.format = 'png'
graph.render("generator_model", format="png")
graph.graph_attr['size'] = '17,17'
display(graph)
```

```python
x = torch.randn(1, 3, 128, 128)
discriminator = Discriminator()
y = discriminator(x)
graph = make_dot(y.mean(), params=dict(discriminator.named_parameters()))
graph.format = 'png'
graph.render("discriminator_model", format="png")
graph.graph_attr['size'] = '17,17'
display(graph)
```
